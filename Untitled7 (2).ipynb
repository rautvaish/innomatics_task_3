```python
#problem1
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the maximum score
max_score = max(employees.values())

# Identify all employees who have that maximum score
top_performers = [name for name, score in employees.items() if score == max_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

```python
#problem2
import string

query = "Buy mobile phone buy phone online"

# Pre-processing: Convert to lowercase and remove punctuation
clean_query = query.lower().translate(str.maketrans('', '', string.punctuation))
words = clean_query.split()

# Count frequencies and filter for those appearing more than once
word_counts = {word: words.count(word) for word in set(words) if words.count(word) > 1}

print(word_counts)
```

```python
#problem3
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Using enumerate to track the hour and filtering for even values
valid_readings = [(index, val) for index, val in enumerate(sensor_readings) if val % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#problem4
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Extract domains and count occurrences
domains = [email.split('@')[1] for email in emails]
domain_counts = {domain: domains.count(domain) for domain in set(domains)}

# Calculate and display percentages
total_emails = len(emails)
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
#problem5
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate the average
avg_sales = sum(sales) / len(sales)
threshold = avg_sales * 1.30  # 30% above average

# Detect and display spikes
for i, amount in enumerate(sales):
    if amount > threshold:
        print(f"Day {i + 1}: {amount}")
```

```python
#problem6
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Use a dictionary to store counts
counts = {uid: user_ids.count(uid) for uid in set(user_ids)}

# Display only those with more than 1 occurrence
for uid, count in counts.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
